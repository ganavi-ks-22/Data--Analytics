```python
df['month_added'] = df['date_added'].dt.month_name()

# Order months properly
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

df['month_added'] = pd.Categorical(df['month_added'], categories=month_order, ordered=True)

# Plot
df['month_added'].value_counts().loc[month_order].plot(kind='bar', color='skyblue')
plt.title("Content Addition by Month")
plt.xlabel("Month")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
country_year = df.dropna(subset=['country', 'year_added']) \
                 .groupby(['year_added', 'country']).size().unstack(fill_value=0)

# Take top 10 countries
top_countries = df['country'].value_counts().head(10).index
plt.figure(figsize=(12,6))
sns.heatmap(country_year[top_countries].T, cmap='YlGnBu', linewidths=1, annot=False)
plt.title("Top Countries by Year (Content Count)")
plt.xlabel("Year")
plt.ylabel("Country")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("netflix_titles.csv")

```

```python
# Convert and clean date column
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df = df.dropna(subset=['date_added'])
df['year_added'] = df['date_added'].dt.year
```

```python
# Split by comma, explode into rows, then count
df['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='barh', color='purple')
plt.title("Top 10 Individual Genres on Netflix")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.tight_layout()
plt.show()

```

```python
df['director'].dropna().str.split(', ').explode().value_counts().head(10).plot(kind='bar', color='orange')
plt.title("Top 10 Directors on Netflix (Cleaned)")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
